Marital Status of customers in terms of Gender
*italicized text*

```python
sns.countplot(df,x='Gender',hue= 'Marital_Status')
```

Marital Status of customers in terms of Gender [Shows Values on Top]

```python
ax = sns.countplot(df,x='Gender',hue='Marital_Status' )
for bars in ax.containers:
  ax.bar_label(bars)
```

Goup by Data in on the basis of Gender and MArital Status and their averag e Amount Spent

```python
df.groupby(['Gender','Marital_Status'],as_index=False)['Amount'].mean().sort_values(by= 'Amount',ascending = False)
```

```python
ab = df.groupby(['Gender'],as_index=False)['Amount'].mean().sort_values(by= 'Amount',ascending = False)
sns.barplot(ab,x='Gender',y= 'Amount')
```

Countplot of age group in terms of Gender

```python
sns.countplot(data = df,x= 'Age Group',hue= 'Gender')
```

Countplot of age group in terms of Gender [Shows Values on top, the value is beiong stored on the containers]

```python
ax = sns.countplot(df,x= 'Age Group',hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending = False)
```

Age group wise sum of amount spending

```python
ab = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending = False)
sns.barplot(ab,x= 'Age Group',y= 'Amount')
```

State wise total order Amount

```python
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending= False)
sns.set(rc={'figure.figsize':(28,10)})
sns.barplot(sales_state,x='State',y='Amount')
```

State wise total orders Count

```python
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by='Orders',ascending= False).head(15)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(sales_state,x='State',y='Orders')
```

```python
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending= False).head(15)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(sales_state,x='State',y='Amount')
```

although two of the last chart almost looks identical , there is a differnce in middle , as one deals is total number of orderes and other shows total order amount
