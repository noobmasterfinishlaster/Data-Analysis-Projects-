```python
covid_df['Active_Case']= covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%Y-%m-%d')
```

**Pivot Table **

```python
statewise = covid_df.pivot_table(values= ['Confirmed','Deaths','Cured'],
                           index ="State/UnionTerritory",aggfunc =max)
statewise
```

Finding Recovery Rate

```python
statewise['RecoveryRAte']= statewise['Cured']*100 / statewise['Confirmed']
```

Finding Cured Rate

```python
statewise['MortalityRate']= statewise['Deaths']*100 / statewise['Confirmed']
```

```python
statewise= statewise.sort_values(by= "RecoveryRAte",ascending = False)
statewise
```

```python
statewise.style.background_gradient(cmap= "cubehelix")
```

top 10 active cases states

```python
top10ActiveCases= covid_df.groupby('State/UnionTerritory')[['Active_Case','Date']].max().sort_values(by= 'Active_Case',ascending = False).reset_index()
top10ActiveCases.head()
```

```python
plt.figure(figsize = (12,8))
plt.xlabel('State')
plt.ylabel('Active_Cases')
plt.title('Top 10 Active Cases')
sns.barplot(top10ActiveCases.iloc[:10],x='State/UnionTerritory', y = 'Active_Case',linewidth = 2 , edgecolor = 'red' )
```
